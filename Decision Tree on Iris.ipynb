```python
import pandas as pd
import requests
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import numpy as np
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
r = requests.get(url, allow_redirects=True)
open('iris.txt','wb').write(r.content)
```

```python
header = ['sepal_length','sepal_width','petal_length','petal_width','names']
df = pd.read_csv('iris.txt',names = header,index_col =False)
df.head(10)
```

```python
X = df.iloc[:,:-1]
X.head(3)
```

```python
y = (df.iloc[:,-1:])
y.head(3)
```

```python
X_train, X_test, y_train, y_test = train_test_split( X,y, test_size = 0.2, random_state = 0)
dtc = DecisionTreeClassifier()
dtc.fit(X_train, np.ravel(y_train))
y_pred = dtc.predict(X_test)
accuracy_score(y_test,y_pred)
```

#### There are 2 major advantage of using decision tree agaisnt other methods, 1 being easy to visualize and understood the ML process and second a certainty of no data scaling requirement. While other ML method may or may not benefits from data scaling, decision tree popularily used without data scaling due to its learning method.

#### While other ml techniques struggle with visualization issue, decision tree took it and done with ease. This leads to one of my technical series which covers the method of visualizing multi-variate ML results of other ML methods

```python
plt.figure(figsize = [16,9]) 
tree.plot_tree(dtc,feature_names = ['sepal_length','sepal_width','petal_length','petal_width'])
plt.show()
```
